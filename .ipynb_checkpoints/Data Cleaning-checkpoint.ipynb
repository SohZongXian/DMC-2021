```python
import pandas as pd
import json
from datetime import datetime
import calendar
import numpy as np
```

```python
ds1 = pd.read_excel('sample_data.xlsx', sheet_name='trans')
ds2 = pd.read_excel('sample_data.xlsx', sheet_name='user')
```

```python
ds1 = ds1.drop(['PriceInfoCurrency','PriceInfoPrice','PriceInfoOldPrice','PriceInfoTax','PriceInfoDiscountType','PriceInfoDiscountValue','PriceInfoDiscountDescription','PriceInfoDiscountStartDate','PriceInfoDiscountEndDate','PriceInfoDiscountDisplayType'], axis = 1)
```

```python
ds1.columns=['attributes', 'categories', 'Item_id', 'merchant_id', 'Item_name','Description', 'order_date', 'order_id', 'price_total', 'user_id']
```

```python
(ds1['order_date'][0]).day_name()
```

```python
ds2.columns = ['user_id', 'age', 'race', 'gender']
```

```python
ds2.nunique()
```

```python
missing_id=list(set(ds1['user_id'])-set(ds2['user_id']))
```

```python
mobile= list(ds2['user_id'])
gender= list(ds2['gender'])
age   = list(ds2['age'])
race  = list(ds2['race'])
mobile = mobile+missing_id
```

```python
ds2 =pd.DataFrame({'user_id':pd.Series(mobile), 'age':pd.Series(age), 'race':pd.Series(race), 'gender':pd.Series(gender)})
```

```python
ds2['age']=ds2['age'].fillna(ds2['age'].mean())
ds2['gender']=ds2['gender'].fillna(ds2['gender'].value_counts().index[0])
ds2['race']=ds2['race'].fillna(ds2['race'].value_counts().index[0])
ds2['age'] = ds2['age'].astype(float).apply(np.ceil)
```

```python
ds2 = ds2.drop_duplicates(subset='user_id', keep='first')
```

```python
ds2=ds2.reset_index()
```

```python
ds2.columns=['UID', 'user','age','race','gender']
```

```python
ds2
```

```python
month =[x.month for x in ds1[ 'order_date']]
day = [x.day for x in ds1[ 'order_date']]
hour = [x.hour for x in ds1[ 'order_date']]
weekday = [x.day_name() for x in ds1[ 'order_date']]
```

```python
ds1=ds1.drop(['order_date'], axis = 1)
```

```python
ds1['month']=month
ds1['day']=day
ds1['hour']=hour
ds1['weekday'] =weekday
```

```python
selected = list(ds1['attributes'])
def selecteditem(x):
    selectpos = x.find("'selected_items':")
    endpos = x.find(",", selectpos)
    return(x[selectpos+17:endpos].replace("'",""))
def forNaN(x):
    if len(x)==1:
        return('NaN')
    else:
        return x
selectedList = [forNaN(x) for x in [selecteditem(x) for x in selected]]
ds1['attributes'] = selectedList
```

```python
ds1.columns
```

```python
ds1.columns = ['selectedItem', 'categories', 'Item_id', 'merchant_id', 'Item_name',
       'Description', 'order_id', 'price_total', 'user','month', 'day', 'hour', 'weekday']
```

```python
df_cleaned = (pd.merge(ds1, ds2, on='user'))
```

```python
df_cleaned.to_csv('cleaned.csv')
```
