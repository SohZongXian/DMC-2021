```python
from scipy.sparse import csr_matrix
#
dsMatrix = df_rating.pivot(index='UID',columns='IID',values='time').fillna(0)
#
item_to_idx ={
    item:i for i, item in 
    enumerate(list(ds.set_index('IID').loc[dsMatrix.index].Item_name))
}

#
mat_item_features_sparse = csr_matrix(dsMatrix.values)
```

```python
###### MATRIX Factorisation#####
R = dsMatrix.as_matrix()
```

```python
user_ratings_mean = np.mean(R, axis = 1)
```

```python
R_demeaned = R - user_ratings_mean.reshape(-1, 1)
```

```python
#user_ratings_mean
```

```python
U, sigma, Vt = svds(R_demeaned, k = 10)
```

```python
U[0]
```

```python
sigma
```

```python
Vt[0]
```

```python
sg= sigma
sigma = np.diag(sigma)
```

```python
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.reshape(-1, 1)
```

```python
preds_df = pd.DataFrame(all_user_predicted_ratings, columns = dsMatrix.columns)
```

```python
def recommend_items(preds_df, userID, item_df, original_ratings_df, num_recommendations=5):
    
    # Get user's predictions
    user_row_number = userID 
    sorted_user_predictions = preds_df.iloc[user_row_number].sort_values(ascending=False) 
    
    # Get the user's data and merge in the item information.
    user_data = original_ratings_df[original_ratings_df.UID == (userID)]
    user_full = (user_data.merge(df_items, how = 'left', left_on = 'IID', right_on = 'IID').
                     sort_values(['time'], ascending=False)
                 )
    # Recommend
    recommendations = (item_df[~item_df['IID'].isin(user_full['IID'])]).merge(pd.DataFrame(sorted_user_predictions).reset_index(), how = 'left', left_on = 'IID',
               right_on = 'IID').rename(columns = {user_row_number: 'Predictions'}).sort_values('Predictions', ascending = False).iloc[:num_recommendations, :-1]
    print("Available transaction of user number: ",userID)
    print('\n',user_full)
    print("\n\nAvailable recommendations: \n")
    print(recommendations)
    return user_full, recommendations
```

```python
user = 1
already_rated, predictions = recommend_items(preds_df, user, df_items, df_rating)
```

```python
def evaluate_SVD(user_row_number,arr):     
    sorted_user_predictions = preds_df.iloc[user_row_number].sort_values(ascending=False) 
    indices = np.array([list(sorted_user_predictions.index.values[1:6])])
    score = np.array([list(sorted_user_predictions.values[1:6].transpose())])
    return arr.append(ndcg_score(indices,score))
```

```python
# filter data to come to an approximation of user likings.
ratings_thres = 1
active_users = list(set(df_users_cnt.query('count >= @ratings_thres').index))
df_ratings_drop_users = df_ratings_drop_item[df_ratings_drop_item.UID.isin(active_users)]
print('shape of ratings data after dropping unpopular items: ', df_ratings_drop_users.shape)
```

```python
# pivot and create item-user matrix
item_user_mat = df_ratings_drop_users.pivot(index='IID', columns='UID', values='time').fillna(0)
#map item titles to index
item_to_idx = {
    item: i for i, item in 
    enumerate(list(df_items.set_index('IID').Item_name))
}
# transform matrix to scipy sparse matrix
item_user_mat_sparse = csr_matrix(item_user_mat.values)
```

```python
# define model
model_knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)
```

```python
def fuzzy_matching(mapper, fav_item, verbose=True):
    
    match_tuple = []
    # get match
    for title, idx in mapper.items():
        ratio = fuzz.ratio(str(title).lower(), fav_item.lower())
        if ratio >= 60:
            match_tuple.append((title, idx, ratio))
    # sort
    match_tuple = sorted(match_tuple, key=lambda x: x[2])[::-1]
    if not match_tuple:
        print('Oops! No match is found')
        return
    if verbose:
        print('Found possible matches in the dataset: {0}\n'.format([x[0] for x in match_tuple]))
    return match_tuple[0][1]
```

```python
def make_recommendation(model_knn, data, mapper, fav_item, n_recommendations):    
    # fit
    model_knn.fit(data)
    # get input item index
    print('You have input item:', fav_item)
    idx = fuzzy_matching(mapper, fav_item, verbose=True)
    
    print('Recommendation system start to make inference')
    print('......\n')
    distances, indices = model_knn.kneighbors(data[idx], n_neighbors=n_recommendations+1)
    
    raw_recommends = \
        sorted(list(zip(indices.squeeze().tolist(), distances.squeeze().tolist())), key=lambda x: x[1])[:0:-1]
    # get reverse mapper
    reverse_mapper = {v: k for k, v in mapper.items()}
    # print recommendations
    print('Recommendations for {}:'.format(fav_item))
    for i, (idx, dist) in enumerate(raw_recommends):
        print('\n{0}: {1}\n with distance of {2}'.format(i+1, reverse_mapper[idx], dist))
```

```python
my_favorite = 'Burger'
make_recommendation(
    model_knn=model_knn,
    data=item_user_mat_sparse,
    fav_item=my_favorite,
    mapper=item_to_idx,
    n_recommendations=5)
```

# KNN Evaluation

```python
def plain_fuzzy_matching(mapper, fav_item, verbose=True):
    
    match_tuple = []
    # get match
    for title, idx in mapper.items():
        ratio = fuzz.ratio(str(title).lower(), fav_item.lower())
        if ratio >= 60:
            match_tuple.append((title, idx, ratio))
    # sort
    match_tuple = sorted(match_tuple, key=lambda x: x[2])[::-1]
    return match_tuple[0][1]
```

```python
def evaluate_recommendation(item,arr):
    model_knn.fit(item_user_mat_sparse)
    idx = plain_fuzzy_matching(item_to_idx, item, verbose=True)
    distances, indices = model_knn.kneighbors(item_user_mat_sparse[idx], n_neighbors=5+1)
    arr.append(ndcg_score(indices, distances))
```

```python
andcgk = []
for each in df_items["Item_name"]:
    evaluate_recommendation(each,andcgk)
np.mean(np.array(andcgk))    
```
